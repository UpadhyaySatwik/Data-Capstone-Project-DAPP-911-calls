```python
# Adding a new column titled "Reason"
title_column = 'title'
df['Reason'] = df[title_column].apply(lambda title: title.split(':')[0].strip())
df
```

```python
# Most common calls 
zip_column = 'Reason'
common_reason = df[zip_column].value_counts()
common_reason
```

```python
# Data Type of Objects in timestamp column
type(df['timeStamp'].iloc[0])
```

```python
# Convert the column from Strings to Date Time Objects
df['timeStamp'] = pd.to_datetime(df['timeStamp'], format='%Y-%m-%d %H:%M:%S')
```

```python
time = df['timeStamp'].iloc[0]
time.hour
```

```python
df['Date'] = df['timeStamp'].dt.to_period('M')
```

```python
month_names = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr',
    5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug',
    9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
```

```python
"Data type of 'timeStamp' column after conversion:", df['timeStamp'].iloc[0]
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
# Create 3 new columns called Hour, Month and Day of Week
df[['Hour', 'Month', 'Day of Week']].head()
```

```python
dmap = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
#Map the actual string names to the day of the week
df[['Hour', 'Month', 'Day of Week']].head()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Day of Week', data=df, hue= 'Reason')
plt.title('911 Calls by Day of Week (Reasons-wise)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1),borderaxespad=0)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Month', data=df, hue='Reason')
plt.title('911 Calls by Day of Week (Reasons-wise)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1),borderaxespad=0)
```

```python
byMonth = df.groupby('Month').count()
```

```python
byMonth.head()
```

```python
plt.figure(figsize=(10, 6))
byMonth['twp'].plot(kind='bar')
plt.title('911 Calls by Month')
```

```python
#Simple plot indicating counts the count of calls per month
plt.figure(figsize=(10, 6))
byMonth['twp'].plot(kind='line')
plt.grid(True)
plt.title('911 Calls by Month')
```

```python
byMonth = byMonth.reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.lmplot(x='Month', y='twp', data=byMonth)
plt.grid(True)
plt.title('Linear Fit of 911 Calls by Month')
```

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())
```

```python
# Plot of counts of 911 calls
plt.figure(figsize=(10, 6))
Date = df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.grid(True)
plt.title('911 Calls per Date')
```

```python
# Plot for Traffic as the reason for 911 calls
df_traffic = df[df['Reason'] == 'Traffic']
df_traffic.groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.grid(True)
plt.title('911 Calls Due to Traffic')
plt.figure(figsize=(10, 6))
```

```python
# Plot for Fire as the reason for 911 calls
df_traffic = df[df['Reason'] == 'Fire']
df_traffic.groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.grid(True)
plt.title('911 Calls Due to Fire')
plt.figure(figsize=(10, 6))
```

```python
# Plot for EMS as the reason for 911 calls
df_traffic = df[df['Reason'] == 'EMS']
df_traffic.groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.grid(True)
plt.title('911 Calls Due to EMS')
plt.figure(figsize=(10, 6))
```

```python
dayHour = df.groupby(['Day of Week','Hour']).count().unstack()['Reason']
dayHour.head()
```

```python
#Heatmap of 911 calls 
plt.figure(figsize=(10, 6))
sns.heatmap(dayHour,cmap='viridis')
plt.tight_layout()
plt.title('911 Calls as per Hours in a week')
```

```python
#Clustermap of 911 calls
plt.figure(figsize=(10,6))
sns.clustermap(dayHour, cmap='viridis')
```

```python
### percentage of 911 calls for each reason
total_calls = len(df)
total_calls
```

```python
df.columns
```

```python
reasons_counts = df['Reason'].value_counts()
```

```python
percentage_by_reason = (reasons_counts / total_calls * 100).round(2)
```

```python
percentage_by_reason
```

```python
plt.figure(figsize=(6, 4))
plt.pie(percentage_by_reason, labels=percentage_by_reason.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of 911 Calls by Reason')
plt.axis('equal')
plt.show()
```
